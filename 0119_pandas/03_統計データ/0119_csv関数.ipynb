```python
# ①Pokemon.csv データを読み込み、先頭から3行を表示する

# pandas のインポート
import pandas as pd

# Pokemon.csvを読み込み、df という変数に格納する。エンコードは utf-8 とする。
df = pd.read_csv("Pokemon.csv")


# dfの先頭から3行を表示する
df.head(3)

```

**# P170～を参考に、それぞれの統計量を求めましょう**

```python
# ヒント：Name列だけを取りだして格納するソース↓
dfname = df.loc[:,'Name']
dfname
```

```python
# Total　列だけを取り出し、Total列内の最大値はいくつか表示してください
dftotal = df.loc[:,"Total"].max()
dftotal
```

```python
# Speed 列だけを取り出し、Speed列内の最小値はいくつか表示してください
dfspeed = df.loc[:, "Speed"].min()
dfspeed
```

```python
# HP列の最頻値、中央値を表示する
dfhp = df.loc[:, "HP"].mode()
print(dfhp)

dfhp = df.loc[:, "HP"].median()
print(dfhp)
```

```python
# 'Legendary'列の内容が True の件数を Count メソッドを使って求め表示してください
# 表示形式がSeriesの形式となって問題ありません。
dflege = df[df.loc[:, "Legendary"] == True ].count()
dflege
```
