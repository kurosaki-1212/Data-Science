```python

# pandas のインポート
import pandas as pd

# yachin.csvを読み込んで、dfという変数に格納する
df = pd.read_csv("yachin.csv", encoding = "shift-jis")

# dfの先頭から10行を表示する
df.head(10)
```

```python
# 統計量をdescribeメソッドをつかって表示
df.describe()
```

```python
# カラム間のデータの相関を確認するため、相関係数を出力する
# 1に近いほど相関が高い(正の相関)
# -1に近いほど相関が高い(負の相関)
# 絶対値が一番1に高いものが相関が高い
df.corr()
```

```python
# 相関係数から、「家賃」列に最も相関が深い列をprintで出力してください。
print("部屋の広さ(m2)")
```

```python
# dfの散布図行列を表示してください。
%matplotlib inline

from pandas.plotting import scatter_matrix
_= scatter_matrix(df)
```

```python
# データが文字化けしてしまったので、列名を変更する。
# 以下のとおり、dfの列名を変更し、df2に格納する。ヒント：P137、昨日の練習問題内の列名変更 columns
# No → No(そのまま)
# 家賃(万円)→ yachin
# 部屋の広さ(m2)→ hirosa
# 駅からの距離(分)→ kyori
# 築年数(年)→ chikunen
# 2階以上→ 2kai
# 南向き→ minami
df2 = pd.DataFrame(df)
df2.columns = ["No", "yachin", "hirosa", "kyori", "chikunen", "2kai", "minami"]
df2
```

```python
# df2の中から'No'列を削除してください
# ヒントP154
df2 = df2.drop("No", axis = 1)
df2
```

```python
# df2の散布図行列を表示してください。
_= scatter_matrix(df2)
```
