CSVデータの読み込み

CSVデータとは…カンマ区切りのデータのこと

```python
import pandas as pd

#例１） ; で区切られている場合の例
student_data_por = pd.read_csv('student-por.csv', sep=';')
student_data_por.head()
```

```python
#練習１）
#  | で区切られた 「ex1.csv」を読み込んでtestdfに格納してみましょう
testdf = pd.read_csv("ex1.csv", sep = "|")
testdf.head()
```

```python
# 例２）
# ヘッダーがないデータは header = None として読み込む必要がある
# ヘッダーを自分で命名する場合は, names=['ヘッダ名1', 'ヘッダ名2'...] とする
# ヘッダ名を命名しない場合は、自動で列番号が付加される
test = pd.read_csv('test.csv', sep='|', header = None)
test
```

```python
# 練習２
# 例２を参考に、ヘッダがない「ex2.csv」データを読み込みましょう。
# ヒント：区切り文字が何か必ず確認してから読み込もう
testdf2 = pd.read_csv("ex2.csv", header=None, names = ["a", "b", "c", "d","e","f","g", "h", "i"])
testdf2
```

**文字コードに応じた読み込み方を覚えよう**

…文字コードが UTF-8 か shift-jis なのかはメモ帳を使いましょう。

```python
# 例３）
# データに日本語が含まれている場合はエンコードの指定が必要
test2 = pd.read_csv('japanese-char.csv', encoding='shift-jis')
test2
```

```python
# 例４）
# UTF-8の場合はエンコードを指定しなくてもOK
test2 = pd.read_csv('japanese-char-2.csv')
test2
```

```python
# 練習３）
# 文字コードがshift-jisのデータ「ex3.csv」を読み込みましょう
testdf3 = pd.read_csv("ex3.csv", encoding = "shift-jis")
testdf3
```
