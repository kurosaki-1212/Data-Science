```python
# pandas のインポート
import pandas as pd

# Pokemon.csvを読み込み、df という変数に格納する。エンコードは utf-8 とする。
df = pd.read_csv("Pokemon.csv")

# dfの先頭から10行を表示する
df.head(10)
```

ダミー変数化とは、「　　数値ではない　　」データを「　　数値　　」データにすること。

**例1）**
['dog','cat','panda']を[0,1,2]に置き換える処理 


**例2) One-Hotエンコーディング**

0119.PNG

このケースは color 列をOne-Hotエンコーディングによるダミー変数化している。

c_はプレフィックスと呼ばれる文字で、ダミー変数化する際に新たに作成される列につけることができる。

```python
# df の'Type 1'列をOne-Hotエンコーディング化しdf_dummyという変数に格納してください。
# プレフィックスは t としてください
df_dummy = pd.get_dummies(df.loc[:, "Type 1"], prefix = "t")
df_dummy
```

```python
# df_dummy のサイズを確認してください
df_dummy.shape
```
