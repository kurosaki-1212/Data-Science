```python
# ①Pokemon.csv データを読み込み、先頭から3行を表示する

# pandas のインポート
import pandas as pd

# Pokemon.csvを読み込み、df という変数に格納する。エンコードは utf-8 とする。
df = pd.read_csv("Pokemon.csv")

# dfの先頭から3行を表示する
df.head(3)

```

```python
# ②joincsv.csv を読み込み、joindf という変数に格納する
joindf = pd.read_csv("joincsv.csv", encoding = "shift-jis")

# joindf の最初から10行を表示する
joindf.head(10)
```

```python
# df の Name 列と joindf の Name 列 を内部結合し、indf というデータに格納する
# 結合の方法は内部結合とする
indf = pd.merge(df, joindf, on = "Name")


# indfの全データを「indf.csv」という名前で保存する
indf
indf.to_csv("indf.csv", encoding = "shift-jis")
```

```python
# df の Name 列と joindf の Name 列 を内部結合し、leftdf というデータに格納する
# 結合の方法は左外部結合とする
leftdf = pd.merge(df, joindf, how = "left")

# indfの全データを「leftdf.csv」という名前で保存する
leftdf
leftdf.to_csv("leftdf.csv")
```
