```python
import numpy as np
import pandas as pd
```

```python
# データ1の準備
data1 = {
    'id': ['100', '101', '102', '103', '104', '106', '108', '110', '111',' 113'],
    'city': ['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo', 'Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
    'birth_year': [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name': ['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve', 'Mituru', 'Aoi', 'Tarou', 'Suguru','Mitsuo']
}
df1 = pd.DataFrame(data1)
df1
```

```python
# データ2の準備
data2 = {
    'id': ['100', '101', '102', '105', '107'],
    'math': [50, 43, 33, 76, 98],
    'english': [90, 30, 20, 50, 30],
    'sex': ['M','F','F','M','M'], 
    'index_num': [0, 1, 2, 3, 4]
}
scoredf = pd.DataFrame(data2)
scoredf
```

**内部結合**

`id`の値が両方の`Dataframe`オブジェクトに存在するものみが表示される。

```python
# データのマージ（内部結合。キーは自動的に認識されるが、onで明示的に指定可能）
# df1 と scoredf を結合。idが同じもの同士で結合したいので、on = 'id' とする
print('・結合テーブル')
pd.merge(df1, scoredf, on='id')
```

**全結合**

どちらのデータにも存在するデータで結合する。
全結合では`how`パラメータに`outer`を指定。結合する値がない場合は、`NaN`になる。

```python
# データのマージ（全結合）
pd.merge(df1, scoredf, how = 'outer')
```

**左外部結合**

左外部結合は`how`パラメータに`left`を指定します。次の例は、左側のテーブル（ひとつめの引数）に合わせて、`Dataframe`オブジェクトのデータを結合する。左側に対応するデータが右（ふたつめの引数）にない場合は、`NaN`になる。

```python
# データのマージ（left）
pd.merge(df1, scoredf, how = 'left')
```

**縦結合**

何らかのキーに紐付いてデータをマージする以外に、`concat`メソッドを使うと、データを縦方向に積み上げられます。→縦結合 P168


```python
# データ3の準備
data3 = {
    'id': ['117', '118', '119', '120', '125'],
    'city': ['Chiba', 'Kanagawa', 'Tokyo', 'Fukuoka', 'Okinawa'],
    'birth_year': [1990, 1989, 1992, 1997, 1982],
    'name': ['Suguru', 'Kouichi', 'Satochi', 'Yukie', 'Akari']
}
df3 = pd.DataFrame(data3)
df3
```

```python
# concat 縦結合 ／concatでdf1とdf3を一つのデータにしている
concat_data = pd.concat([df1,df3])
concat_data
```
