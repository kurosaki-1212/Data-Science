```python
import numpy as np

# numpyのarange関数を使って0～10までの配列 a1 を作成し、printを使って表示する
a1 = np.arange(10)
print("a1:", a1)

# numpyのarange関数を使って10～120まで、10ずつ増加する値が代入された配列 a2 を作成し、printを使って表示する
a2 = np.arange(10, 120+1, 10)
print("a2:", a2)

# numpyのarange関数を使って0～11までの配列を作成し、reshapeを使って2×6の二次元配列に変換し変数 a3 に代入し、a3の内容をprintで表示する。
a3 = np.arange(12)
a3 = a3.reshape(2, 6)
print("a3:", a3)
```

**P104：乱数**

**【ポイント】**

**空欄は授業をよく聞いてメモっておきましょう。**

・乱数とは、規則性のないバラバラな数のことを言う。

・乱数のseed（シード）値とは、本来いつも異なる数を発生させる乱数の値を、固定する設定値

・乱数のseed値を指定すると、何度実行しても[     ①    ]が得られることを担保できる。

・一様乱数とは[     ②      ]のこと。**一様分布**とも言う。
　※出やすい値や出にくい値がない、という状態

・正規分布乱数とは、正規分布に従って、分布の中心に行くほど出現する確率が高くなる乱数のこと。

【答え】
①同じ乱数の結果
②数値が同じ確率で出現する、 確率分布に従う乱数

```python
# 乱数シード値を10に設定する
np.random.seed(10)

# ② 1以上、100未満の一様分布の乱数を20個発生させて、4行5列の行列を生成し変数m1に代入する
m1 = np.random.randint(1, 100, (4, 5))

# ③ -100 以上、100未満の一様分布の乱数を20個発生させて、4行5列の行列を生成し変数m2に代入する
np.random.seed(10)
m2 = np.random.randint(-100, 100, (4, 5))

# ④m1、m2をそれぞれprintで表示する
print(m1)
print(m2)

# ⑤concatinate関数を使って、m1とm2を列方向に連結し、m3に代入する
m3 = np.concatenate([m1, m2], axis = 1)


# ⑥ m3をprintで表示する
print(m3)
```

```python
# ⑦concatinate関数を使って、m1とm2を行方向に連結し、m4に代入する
m4 = np.concatenate([m1, m2], axis = 0)

# ⑥ m4をprintで表示する
print(m4)
```

P132：関数とメソッド

配列内の合計、平均、最大値、最小値を関数で求めることができる。

```python
# ①以前作成した m4 を再度printで表示する。
print(m4)

# ②m4の全体の合計値をsum関数を使用して求めprintで表示する
print("合計値",  np.sum(m4))

# ③m4の0列目の最大値を求め、printで表示する
print("最大値", np.max(m4[0]))

# ④m4の2行目の最小値を求め、printで表示する
print("最小値", np.min(m4[2]))

print("")

num = np.arange(1, 15 + 1)
num = num.reshape(5, 3)

print(num)
print("合計値", np.sum(num[:, 0]))
print("最小値", np.min(num[2:4, 1:]))
print("最大値", np.max(num))
```

## 練習問題②

標準正規分布に従う乱数を10個発生させて配列を作成してください。その中の最大値、最小値、合計値を求めて表示してください。
乱数のシード値は100とします。

```python
np.random.seed(100)
num = np.random.random(10)
print("最大値", max(num))
print("最小値", min(num))
print("合計値", sum(num))
```
