car_sales.csv を読み込んで、以下のコメントのとおり、学習データとテストデータに分割するプログラムを実装してください。

```python
# pandas をインポートし、pd という名前で使えるようにする
import pandas as pd

# car_sales.csvを読み込み、df という変数に格納する。
df = pd.read_csv("car_sales.csv")

# df のデータの末尾5件を表示する
df.tail()
```

```python
# 読み込んだcsvデータの行数、列数を表示する
df.shape
```

```python
# 読み込んだcsvデータの欠損値の件数を表示する
df.isnull().sum()
```

```python
# 説明変数の列（Price_in_thousands から Power_perf_factor までの列 ）のみ取り出して、Xに代入する
X = df.iloc[:, :-1]
# Xの末尾から5行を表示し、適切な列が取得できているか確認する
X.tail()
```

```python
# Xの列数、行数を確認する
X.shape
```

```python
# 目的変数の列（=Sales_in_thousands 列）のみ取り出して、yに代入する
y = df.iloc[:, -1]
# yの末尾から5行を表示し、適切な列が取得できているか確認する
y.tail()
```

```python
# yの列宇数、行数を確認する
y.shape
```

```python
# データ分割（学習データとテストデータ）のための train_test_split 関数のインポート
from sklearn.model_selection import train_test_split

# 学習データとテストデータに分ける
# 分割は 70%をテストデータにする
# random_stateは、10を設定すること
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.7, random_state = 10)
```
