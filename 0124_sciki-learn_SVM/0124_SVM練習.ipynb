```python
import pandas as pd
import sklearn
from sklearn.svm import LinearSVC

cancer = pd.read_csv("cancerdata-Nanあり.csv")
cancer.head()
```

```python
cancer = cancer.fillna(method="ffill")
```

```python
print('データの欠損数:',cancer.isnull().sum().sum())
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

y = cancer["diagnosis"]
y
```

```python
X = cancer.drop(["diagnosis"], axis = 1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.09, random_state=600)
```

```python
model = LinearSVC()

model.fit(X_train, y_train)

train_score = model.score(X_train, y_train)

test_score = model.score(X_test, y_test)

print("学習データ結果",train_score)
print("テストデータ結果", test_score)
```
