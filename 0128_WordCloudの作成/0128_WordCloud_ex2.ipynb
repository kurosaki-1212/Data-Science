テキスト「100本ノック」の例を簡略にし、データクラウドを表示する例を以下に示します。

```python
# MeCabのインストール
import MeCab

tagger = MeCab.Tagger()

# tagger.parse(【対象の文字列】)
# 対象の文章の形態素解析を行う。（日本語を最小単位の品詞に分解する）
parsed = tagger.parse(content).split('\n')

# 形態素解析の結果である「parsed」の出力結果を確認する
# ※先頭の10行のみ表示
parsed[:10]
```

```python
# 上記のデータのように形態素解析の結果は、1行の中の\t(タブ文字)の左側のみが単語データです。
# 一行の中を\tで区切って、単語データのみをWordCloudの元データとします。
import re

# parsedの中のデータを1行ずつ読み取り、「\t」か「,」でsplitする
items = (re.split('[\t,]', line) for line in parsed)

words = []
for item in items:
  if item[0] == 'EOS' or item[0] == '':
    pass
  elif item[1] in ["名詞", "形容詞", "動詞", "副詞"]:
    # 助詞、助動詞は含めない
    words.append(item[0])

# textがワードクラウド生成の対象
text = ' '.join(words)
```

ワードクラウドを描画する準備を行います。

```python
# WordCloudを表示する準備
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# フォントをインストールしたパス（ここは変更しなくてよい）
font_path = 'usr/share/fonts/opentype/ipafont-gothic/ipagp.ttf'

# ここでワードクラウドの元データを生成している
cloud = WordCloud(background_color='white', font_path=font_path).generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(cloud)
plt.axis("off")
# ファイル名は適宜変更してください
plt.savefig('wc_noun_base_2.png')
plt.show()
```

③別の方法でもWordCloudを生成してみましょう。
今回はparseToNodeを使ってテキストを分解します。

parseToNodeで生成した戻り値nodeは, node.nextで次のデータを呼び出せる優れものです。
単語データはnode.surfaceに格納されています。

```python
# MeCabの準備
tagger = MeCab.Tagger()
tagger.parse('')
node = tagger.parseToNode(content)
```

```python

# 名詞、形容詞だけを取り出す
word_list = []
while node:
  word_type = node.feature.split(',')[0]
  if word_type == '名詞':
    word_list.append(node.surface)
  elif word_type == '形容詞':
    word_list.append(node.surface)
  node = node.next

# リストを文字列に変換
targetdata = ' '.join(word_list)

# ここでワードクラウドの元データを生成している
cloud = WordCloud(background_color='white', font_path=font_path).generate(targetdata)

plt.figure(figsize=(10, 5))
plt.imshow(cloud)
plt.axis("off")
# ファイル名は適宜変更してください
plt.savefig('wc_noun_base_2.png')
plt.show()
```
