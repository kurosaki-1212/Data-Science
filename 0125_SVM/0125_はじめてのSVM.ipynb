# **「cancerdata-Nanあり.csv」を読み込んで、SVM のモデルを作成しましょう。**


目標：テストデータのスコアが75%以上となるよう、データの前処理を行うこと！ 

```python
# pandasのライブラリ
import pandas as pd
# 機械学習ライブラリ
import sklearn

# SVMのライブラリを読み込む
from sklearn.svm import LinearSVC

# データの読み込み
cancer = pd.read_csv('cancerdata-Nanあり.csv')

# データのデータ内容を確認
cancer.head()
```

```python
# 列ごとの欠損値の確認
cancer.isnull().sum()
```

```python
# 欠損値のデータをdrop してみる P164
cancer_drop = cancer.dropna()
```

```python
# 欠損値の確認
cancer_drop.isnull().sum()
```

```python
# SVMのライブラリの読み込み
from sklearn.svm import LinearSVC
# 学習データとテストデータを分けるライブラリの読み込み
from sklearn.model_selection import train_test_split

# yに目的変数：「diagnosis」の列を代入する
y = cancer_drop['diagnosis']
# yに何が入っているか確認する
y
```

```python
# Xに説明変数の列を代入する
#  以下は「diagnosis」列を削除して、それ以外のデータをyに代入している axis=1→列方向
X = cancer_drop.drop(['diagnosis'], axis=1)

# Xに何が入っているか確認
X.head()
```

```python
# 学習データとテストデータに分割する
# test_size = 0.5 ,random_state = 10 とする
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=10)
```

```python
# モデルの初期化、学習を行う
model = LinearSVC()

# 学習データを使った学習（ fit : trainデータを使って、学習）
model.fit(X_train, y_train)

trainscore = model.score(X_train, y_train)
# 作成したモデルに対し、訓練データを使って精度を確かめる
test_score = model.score(X_test, y_test)

print('学習データの学習結果：',trainscore)
print('テストデータの学習結果：',test_score)
```

```python
# スコアが悪すぎるので、目的変数との相関を調べることに。
cancer_drop.corr()
```

```python
# 'id','fractal_dimension_mean'は明らかに相関係数が低い
# 'id'を削除
cancer_drop2 = cancer_drop.drop(['id'], axis=1)
# 削除されているか確認
cancer_drop2
```

```python
# fractal_dimension_mean 列も削除
cancer_drop3 = cancer_drop2.drop(['fractal_dimension_mean'], axis=1)
# 削除されているか確認
cancer_drop3
```

```python
# 不要そうな列を削除したので、再度、説明変数の列を作成
# Xに説明変数の列を代入する
#  以下は「diagnosis」列を削除して、それ以外のデータをyに代入している axis=1→列方向
X = cancer_drop3.drop(['diagnosis'], axis=1)

# Xに何が入っているか確認
X.head()
```

```python
# もう一度学習データとテストデータに分割する
# 今回は、test_size = 0.3 ,random_state = 10 とする
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=10)
```

```python
# モデルの初期化、学習を行う
model = LinearSVC()

# 学習データを使った学習（ fit : trainデータを使って、学習）
model.fit(X_train, y_train)

trainscore = model.score(X_train, y_train)
# 作成したモデルに対し、訓練データを使って精度を確かめる
test_score = model.score(X_test, y_test)

print('学習データの学習結果：',trainscore)
print('テストデータの学習結果：',test_score)
```
