# P177 オブジェクト指向スタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# 折れ線グラフを描画
ax.plot(x, y)

# タイトルを設定
ax.set_title("OOP-style")

# グラフを表示
plt.show()  
```

例）
以下は家賃データを読み込んで、家賃と家の広さを散布図として表示する例です。

散布図の作成方法はテキスト P189～190 を確認しながら進めましょう。


```python
# pandas のインポート
import pandas as pd

# yachin.csvを読み込んで、dfという変数に格納する
df = pd.read_csv("yachin.csv")

# dfの先頭から5行を表示する
df.head()
```

```python
# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# x軸（横軸）に家賃、y軸（縦軸）に広さを設定、色を赤にする
ax.scatter(df['yachin'], df['hirosa'], c="red")

# x軸（横軸）のラベルを設定
ax.set_xlabel('yachin')

# y軸（縦軸）のラベルを設定
ax.set_ylabel('hirosa')

# 表示
plt.show()
```

**練習）**

**上の例を参考に、以下の散布図を作成してください。**

①グラフスタイルに ggplot を指定

②x軸に家賃、y軸に駅からの距離（列名：eki）を設定する

③プロットする色を green にする

④マーカーの形は上向き三角▲とする（P190）

⑤x軸（横軸）のラベルに「yachin」を設定

⑥y軸（縦軸）のラベルに「eki」を設定

⑦グラフタイトルに「graph0120」を設定（P178）

```python
# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# ①グラフのスタイルに ggplot を指定
matplotlib.style.use("ggplot")

# ②、③、④ x軸に家賃、y軸に駅を設定し、色をgreenにし、▲でプロットする
ax.scatter(df["yachin"], df["eki"], c = "green", marker = "^")

# ⑤x軸（横軸）のラベルを設定
ax.set_xlabel("yachin")

# ⑥y軸（縦軸）のラベルを設定
ax.set_ylabel("ekia")

# ⑦グラフタイトルの設定
ax.set_title("graph0120")

# 表示
plt.show()
```
