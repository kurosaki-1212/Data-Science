```python
# Matplotlibのインポート
import matplotlib.pyplot as plt

# matplotlib.styleのインポート
import matplotlib.style

# ggplotスタイルを設定
matplotlib.style.use('ggplot')
```

```python
# 元となるデータの読み込み

# pandas のインポート
import pandas as pd

# catsdogs.csvを読み込んで、dfという変数に格納する
df = pd.read_csv('catsdogs.csv')

# dfの先頭から5行を表示する
df.head()
```

```python
# 箱ひげ図の描画の対象データ「犬」の列を dogsdf に格納
dogsdf = df.loc[:, 'dogs']

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# グラフタイトル
ax.set_title('dogs weight')

# 表示ラベルを設定
labelsdata = ['dogs']

# 箱ひげ図を描写
ax.boxplot(dogsdf,labels=labelsdata)

# 表示
plt.show()
```

上のグラフで、70の位置に白い丸が描画されています。これは何というか、print で表示してください。

上の例を参考に、猫の箱ひげ図を犬の箱ひげ図の横に並べる形で描画しましょう。


```python
# 箱ひげ図の描画の対象データ「犬」の列を dogsdf に格納
dogsdf = df.loc[:, "dogs"]

# 箱ひげ図に描画するための猫の列を catsdf に格納
catsdf = df.loc[:, "cats"]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# グラフタイトル
ax.set_title("cats and dogs weight")

# 表示ラベルを設定
labels = ["cats", "dogs"]

# 箱ひげ図を描写
ax.boxplot((catsdf, dogsdf),labels=labels)

# 表示
plt.show()
```

```python
# グリッドを表示せず、横向きに表示する例

# 箱ひげ図の描画の対象データ「犬」の列を dogsdf に格納
dogsdf = df.loc[:, 'dogs']

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# グラフタイトル
ax.set_title('dogs weight')

# 表示ラベルを設定
labelsdata = ['dogs']

# 箱ひげ図を描写  vert=False で横向きになる
ax.boxplot(dogsdf,labels=labelsdata, vert=False)

# グリッドを表示しない
ax.grid(False)

# 表示
plt.show()
```
