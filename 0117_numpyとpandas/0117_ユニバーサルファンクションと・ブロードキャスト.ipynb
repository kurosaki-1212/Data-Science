ユニバーサル関数とも言う

https://numpy.org/devdocs/user/basics.ufuncs.html


```python
import numpy as np
```

```python
# ユニバーサル関数
sample_data = np.arange(10)
print('元のデータ：', sample_data)
print('すべての要素の平方根：',np.sqrt(sample_data))
print('すべての要素のネイピア指数関数：',np.exp(sample_data))
```

**P122 ブロードキャスト**

配列の大きさが異なっているときに、自動的に要素をコピーして、対象の大きさを揃える機能
以下サンプルで動作を確認しよう

```python
sample_array = np.arange(10)

print(sample_array)
```

以下の例がブロードキャスト。
sample_arrayは配列で、3は配列ではないので、本来はそのまま計算できない。
Numpyでは暗黙的に、要素をコピーして大きさを揃えて、`sample_array + np.array([3,3,3,3,3,3,3,3,3,3])`のように計算してくれる。
**これがブロードキャストです。**

```python
result = sample_array + 3
print(result)
```
