```python
import pandas as pd
```

```python
# Excelを読み込む
df = pd.read_csv("Pokemon.csv", encoding= "utf-8")

# 上から表示するにはhead関数を使う
df.head()
```

```python
# 範囲指定もできる
df.iloc[0:3 ,0:4]
```

```python
# 後ろから表zするにはtail関数を使う
df.tail()
```

```python
# shapeで行数と列数が手に入る
psize = df.shape
print("行数は" , psize[0])
print("列数は" , psize[1])
```

```python
# columnsとindexはvaluesで表示できる
df.columns.values
```

```python
# culumnsとindexは名前の変更もできる
df1 = df.copy()
df1.columns = ['#', '名前', 'Type 1', 'Type 2', 'Total', 'HP', 'こうげき',
       'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation',
       '伝説のポケモンか']
df1.head()
```

```python
# DataFrameの作り方
import numpy as np

num = np.array([45, 47, 8, 56, 41, 3])
name_df = pd.DataFrame(num.reshape(2, 3),
                        columns = ["きのこ", "たけのこ", "どちらでもない"],
                        index = ["愛知県", "東京都"])
name_df
```

```python
# DataFrameの別の作り方
pd.DataFrame({"きのこ": [45, 56], "たけのこ": [47, 41]})
```

```python
# columnsやindexを指定してその行列を表示できる
namedf = df.loc[:, ["Name", "Attack"]]
namedf.head(3)
```

```python
# 条件で抽出
hp55df = df[df["HP"] >= 55]
hp55df.head(3)
```

```python
# queryを使いcolumnsの中の変数名を指定して表示できる
pikadf = df.query("Name == 'Pikachu'")
pikadf
```

```python
# 並び替え(基本昇順)
# ascending = Falseで降順にできる
total = df.sort_values(by = "Total", ascending = False)
total.head()
```

```python
# 不要なカラムの削除
df2 = df.drop("#", axis = 1)
df2.tail()
```
